```python
##Testing purpose
from IPython.display import display

user_prompt = []
# Function to handle the chat interaction
def handle_chat():
    user_message = input("User: ")
    user_prompt.append(user_message)

    # Getting response from GPT
    ai_response = generate_chat_completion(user_message)
    if ai_response is None:
        print("ChatBot: Sorry, I'm having trouble processing your request. Please try again.")
        return

    if "top_pics" in ai_response:
        print("Based on your profile, we recommend you the following stocks:")
        top_picks = get_top_pro_pics()
        records = []
        for item in top_picks['data']:
            record = {
                'symbol': item['symbol'],
                'name': item['name'],
                'public Tags': item['public_tags'],
                'description': item['tag_info']['Description'],
                'type': item['type'],
            }
            records.append(record)
        df = pd.DataFrame(records)
        df = df[df['public Tags'].apply(len) > 0]
        df['tags_length'] = df['public Tags'].apply(len)
        df = df.sort_values(by='tags_length', ascending=False)
        df = df.drop(columns='tags_length')
        df.reset_index(drop=True, inplace=True)
        display(df)
        
        user_input = input("Would you like check to performance of any stock? (Yes/No): ").strip().lower()
        
        #scenario_handle(f'{user_input}')
        if user_input.lower() == 'yes':
            print("Which stock performance you would like to see")
            handle_chat()          
        elif user_input.lower() == 'no':
            print("ChatBot: Thank you! Have a great day!")
            return False
        else:
            print("Please try again")
            handle_chat()
        
    elif "ticker:" in ai_response:
        stock_ticker = ai_response.split(" ")[1]
        print(f"Stock information for {stock_ticker}:")
        stock_info = get_stock_info(stock_ticker)
        columns = ['Prev Close', 'Open', 'Current', 'Low', 'High', '52 wk low', '52 wk high']
        values = []
        for col in columns:
            for item in stock_info['data']:
                if item['Name'] == col:
                    values.append(item.get('Double', item.get('Int64')))
        df = pd.DataFrame([values], columns=columns)
        print(f"Current Price: {df['Current'][0]}")
        print(f"Open at: {df['Open'][0]},    Prev Close : {df['Prev Close'][0]}")
        print(f"Todays Low: {df['Low'][0]},  Todays High: {df['High'][0]}")
        print(f"52weeks Low: {df['52 wk low'][0]},  52weeks  High: {df['52 wk high'][0]}")
        print("Note: All the values are in USD")
        
        
        user_input = input("Would you like to see more stocks? (Yes/No): ").strip().lower()
        if user_input.lower() == 'yes':
            print("Which stock performance you would like to see ?")
            handle_chat()
            
        elif user_input.lower() == 'no':
            print("ChatBot: Thank you! Have a great day!")
            return False
        else:
            print("Please try again")
            handle_chat()
            
    else:
        print(f"ChatBot: {ai_response}")
        #print("ChatBot: I'm sorry, I can only assist with stock recommendations and stock information as of now")
    

# Main function to start the chatbot interaction
def main():
    print("You can start the conversation by typing 'Recommend me something'.")
    while True:
        handle_chat()
```
